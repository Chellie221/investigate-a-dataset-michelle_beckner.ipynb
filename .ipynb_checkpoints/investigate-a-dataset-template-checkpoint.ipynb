### Data Cleaning (Replace this with more specific notes!)

```python
# I want to remove any row that has a NaN. I want to use only countries that have data for all the years since 
# I'm going to calculate the mean of each country and take the top 5 and bottom 5. I'm curious to see if the 
# comparisons are any different between the top and bottom of the life expectancy range.

life_expect = pd.DataFrame(life_expect).dropna(axis=0)
```

```python
# I know for more of an accurante measure, I would need the full float. However, since I'm just wanting to look 
# at a quick correlation, I want to use whole numbers for ease of reading. 

pd.set_option('display.float_format','{:.0f}'.format)
```

```python
# I want to find the mean life expectancy of each country to begin the process of choosing the the countries 
# I want to look at. 

life_expect['LifeExpectancy'] = life_expect.iloc[:, 1:].mean(axis=1)
```

```python
# The dataset contains data for years 1799 â€“ 2099.  I need to pair the time frame down to make it more management. 
# I choose to use the 20th century as the time frame. I choose this period because I feel the data will be more 
# reliable. The bottom part of the data set and the top part of the dataset are based on estimations, trends and 
# formulas. I then decided to use each decade at a point to see the data throughout the century. I created a 
# dataset that help just the years that I wanted to look at.

life_year_subset=life_expect[['country', '1900', '1910', '1920', '1930', '1940', '1950', '1960', '1970', '1980', '1990', '2000', 'LifeExpectancy']].sort_values('LifeExpectancy', ascending=False)
```

```python
# I want to find the 5 countries with the highest mean.

life_highest_five = life_year_subset[:5]
```

```python
# I want to find the 5 countries with the lowest mean.

life_lowest_five = life_year_subset[-5:]
```

```python
# I'm going to merge the 5 highest and 5 lowests lists to make one list of the countries I want to look at.

life_final = pd.merge(life_highest_five, life_lowest_five, how='outer').sort_values('LifeExpectancy', ascending=False)
```

```python
# I'm going to drop the years and just keep the mean column. I will use this dataset to add columns for the mean column 
# of the other 2 datasets so that I have one column to analyze. I will create a function to do this since I have to
# repeat this task.

def remove_years(filename):
    filename = filename.drop(['1900', '1910', '1920', '1930', '1940', '1950', '1960', '1970', '1980', '1990', '2000'], axis=1, inplace=True)
```

```python
remove_years(life_final)
```

```python
# I'm going to merge the life_final dataset with the children dataset so that the file contains the 10 countries
# we selected to review with the children per woman data. 

children_final = pd.merge(life_final, children, how='left', left_on='country', right_on='country')
```

```python
children_final = children_final[['country', '1900', '1910', '1920', '1930', '1940', '1950', '1960', '1970', '1980', '1990', '2000']]
```

```python
# I'm going to find the mean of each country.

children_final['NumofChildren'] = children_final.iloc[:, 1:].mean(axis=1)
```

```python
# Now it's time to drop the years and just keep the mean column.

remove_years(children_final)
```

```python
# I'm now going to repeat this same process for the child mortality dataset.

mortality_final = pd.merge(life_final, child_mortality, how='left', left_on='country', right_on='country')
```

```python
mortality_final = mortality_final[['country', '1900', '1910', '1920', '1930', '1940', '1950', '1960', '1970', '1980', '1990', '2000']]
```

```python
mortality_final['ChildMortality'] = mortality_final.iloc[:, 1:].mean(axis=1)
```

```python
remove_years(mortality_final)
```

```python
# I'm going to merge all 3 datasets now to make one dataset with the information I want to compare.

summary = life_final.merge(children_final,on='country').merge(mortality_final,on='country')
print(summary)
```

```python
summary.describe()
```

```python
summary.hist()
```
